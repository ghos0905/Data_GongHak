```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
```

```python
# 데이터 이름 바꿧음 11-22 어쩌구저꺼죽
data = pd.read_csv("소득반영최종.csv")
```

```python
columns = ["소득(동별)","인구밀도","사업체수","장애등급별 장애인현황","독거노인수","노령화지수"]
```

```python
data.columns
```

```python
xtrain = data[columns].rename(columns={"장애등급별 장애인현황":"장애등급"})
```

```python
SNR, DMJ, GM = data.새누리당, data.더불어민주당, data.국민의당
```

```python
RF_SNR = RandomForestRegressor().fit(xtrain,SNR)
```

```python
RF_DMJ = RandomForestRegressor().fit(xtrain,DMJ)
```

```python
RF_GM = RandomForestRegressor().fit(xtrain,GM)
```

```python
test = pd.read_csv("최근년도_predict_data_final_1.csv")
```

```python
model = test.drop(["읍면동명","자치구"],axis=1).dropna()
```

```python
SNR_pred = RF_SNR.predict(model)
```

```python
DMJ_pred = RF_DMJ.predict(model)
```

```python
GM_pred = RF_GM.predict(model)
```

```python
SNR_pred
```

```python
test
```

```python
final = test.dropna()[['읍면동명','자치구']].assign(새누리당 = SNR_pred, 더민주=DMJ_pred)
#final = final.groupby("자치구").sum()
#final = final.assign(SNR=final.새누리당 > final.더민주, DMJ=final.새누리당 < final.더민주)
```

```python
final_predict=pd.merge(test, final, how="left")
```

```python
final_predict.to_csv("final_predict")
```
