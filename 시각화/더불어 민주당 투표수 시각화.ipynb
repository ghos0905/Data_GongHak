```python
# data handling
import pandas as pd
import numpy as np
import json
import geopandas as gpd
import re

# map vis
import folium
from folium import plugins

data1 = pd.read_csv("소득반영최종.csv").drop(columns='Unnamed: 0')
data1 = data1[data1.year == 2016]
```

```python
geodf = gpd.read_file('HangJeongDong_ver20160201geojson.json')
```

```python
geodf=geodf.dropna()
```

```python
geodf["adm_nm"]=geodf.adm_nm[geodf.adm_nm.str.contains(pat = '서울특별시')]
```

```python
geodf=geodf.dropna()
```

```python
geodf
```

```python
geodf = geodf.astype({"adm_cd":"int"})
codes= pd.read_excel("한국행정구역분류_2016.07.01기준.xls", header=2)
codes
```

```python
codes=codes.iloc[2:450, 4:7]
codes
```

```python
codes=codes.dropna()
```

```python
codes
```

```python
codes=codes.astype({"소분류":"int"})
```

```python
codes
```

```python
codes=codes.rename(columns=({"소분류":"읍면동코드"}))
```

```python
codes=codes.rename(columns=({"읍면동":"읍면동명"}))
```

```python
codes=codes.rename(columns=({"시군구":"자치구"}))
```

```python
codes
```

```python
geodf=geodf.astype({"adm_cd":"int"})
geodf
```

```python
geodf=geodf[geodf.adm_cd.isin(codes.읍면동코드)]
```

```python
geodf
```

```python
geodf=geodf.rename(columns={'adm_cd':'읍면동코드'})
```

```python
geodf=geodf.reset_index().drop("index", axis=1)
```

```python
geodf
```

```python
codes=codes.reset_index().drop("index", axis=1)
```

```python
codes
```

```python
len(set(geodf.읍면동코드) & set(codes.읍면동코드))
```

```python
data= pd.merge(geodf, codes, how="left")
data
```

```python
data=data.drop(["adm_nm", "geometry"], axis=1)
data
```

```python
data1 = data1[["자치구",'읍면동명','더불어민주당']]
data1
```

```python
data["읍면동명"]=data.읍면동명.apply(lambda x : re.sub("·", ".", x))
```

```python
data["읍면동명"]=data.읍면동명.apply(lambda x : re.sub("종로5.6가동", "종로5·6가동", x))
```

```python
data1[data1.자치구=="중구"]
```

```python
data[data.자치구=="성동구"]=data[data.자치구=="성동구"].replace("성동구", "중구")
```

```python
data[data.자치구=="중구"]
```

```python
data1[data1.읍면동명.isin(data.읍면동명)]
```

```python
data=pd.merge(data1, data, how="left")
```

```python
data=data.astype({'읍면동코드':'int'})
data
```

```python
tmp = pd.merge(geodf, data, how="left")
tmp
```

```python
tmp=tmp.drop("adm_nm", axis=1)
```

```python
data=pd.merge(tmp, data, how="left")
```

```python
tmp=tmp.drop("자치구", axis=1)
```

```python
tmp["읍면동명칭"]=tmp["읍면동명"]
```

```python
tmp=tmp.set_index("읍면동명칭")
```

```python
tmp
```

```python
data=data.drop("자치구", axis=1)
```

```python
data
```

```python
geojson = json.loads(tmp.to_json())

googlemap = folium.Map(location=[37.5502, 126.982], zoom_start=10.5)

choropleth = folium.Choropleth(
    geojson,
    data=data,
    key_on='feature.id',
    columns=['읍면동명', '더불어민주당'],
    fill_color='RdYlGn',
    fill_opacity=0.5).add_to(googlemap)

choropleth.geojson.add_child(
    folium.features.GeoJsonTooltip(['읍면동명','더불어민주당'])
)
```
