```python
#라이브러리 호출
import numpy as np
import pandas as pd

from sklearn.ensemble import RandomForestRegressor , GradientBoostingRegressor
from sklearn.metrics import mean_squared_error

from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
```

```python
#원본 데이터 (깃허브에 저장된 데이터 파일 호출)
df1 = pd.read_csv('https://github.com/lhmlhm1111/Data_GongHak/raw/master/%EB%AA%A8%EB%8D%B8/12%2C16%EB%85%84%EB%8F%84_Training_Data_set_1.csv',encoding='utf-8',sep=',')

final = pd.read_csv('https://github.com/lhmlhm1111/Data_GongHak/raw/master/%EB%AA%A8%EB%8D%B8/%EC%B5%9C%EA%B7%BC%EB%85%84%EB%8F%84_predict_data_final_1.csv',encoding='UTF-8',sep=',')
```

```python
# 회의 때 변수 선별 3가지 방법 univariate, importance , heat map상관관계도 로 선별한 6가지
# 소득 , 인구밀도 , 사업체수 , 장애등급별 장애인현황 ,독거노인수 , 노령화지수
# 로 Train , Test 머신러닝 한 과정입니다
```

```python
#학습/테스트 데이터
x = df1.drop(['year','읍면동명','자치구','새누리당','더불어민주당','국민의당','선거인수','유년부양비','노년부양비','종사자수','사업체 창업률','기초수급자인원수','보육시설','금융기관','사업체 평균연령','인구','면적'],axis=1)
y = df1['새누리당']

x_train, x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
## 랜덤포레스트 회귀 모델 ( 새누리당 예측 ! )

rf1 = RandomForestRegressor()

rf1.fit(x_train, y_train)
```

```python
a=['인구밀도','사업체수','장애등급','독거노인수','노령화지수']
for i in range(5):
    final[a[i]]=final[a[i]].fillna(final[a[i]].median())
```

```python
final_x = final.drop(final.columns[0:2],axis=1)
```

```python
predict1 = rf1.predict(final_x)
```

```python
df4 = pd.DataFrame(predict1,columns=['새누리당'])
```

```python
df5 = final.drop(final.columns[2:],axis=1)
```

```python
final1 = pd.concat([df5,df4],axis=1)
```
