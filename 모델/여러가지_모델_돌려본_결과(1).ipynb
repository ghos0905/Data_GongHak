```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor,VotingRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn import metrics
from sklearn.model_selection import KFold, cross_val_score
```

```python
X, y = election_total.iloc[:,7:], election_total.iloc[:,4]
cv = KFold(10,shuffle=True)
def eval_model(model, x=X, y=y, cv=cv):
    score = cross_val_score(model, x, y, cv=cv, scoring='neg_mean_squared_error') *-1
    score = np.sqrt(score)
    return score.mean()

import warnings
warnings.filterwarnings("ignore")
```

```python
LR = LinearRegression()
LASSO = Lasso()
RIDGE = Ridge()
SV = SVR()
RF = RandomForestRegressor()
ADA = AdaBoostRegressor()
GB = GradientBoostingRegressor()
models = [LR ,LASSO, RIDGE, SV, RF, ADA, GB]
names=['Linear', 'Lasso', 'Ridge', 'SupportVector','RandomForest', 'AdaBoost','GradientBoost']
```

```python
VR = VotingRegressor([("Linear", LASSO), ("Rf", RF), ("AdaBoostr", ADA),("Gradient", GB)])
```

```python
eval_model(VR)
```

```python
for model, name in zip(models, names):
    print(name)
    print(eval_model(model))
    print("===")
```
