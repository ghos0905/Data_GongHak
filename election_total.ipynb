```python
import pandas as pd
```

```python
election_2012=pd.read_csv("19대 총선 전처리(2).csv")
election_2012=election_2012.drop("Unnamed: 0", axis=1)
```

```python
election_2012["선거인수"]=election_2012["선거인수"].apply(lambda x : str(x))
election_2012["새누리당"]=election_2012["새누리당"].apply(lambda x : str(x))
election_2012["민주통합당"]=election_2012["민주통합당"].apply(lambda x : str(x))
election_2012["통합진보당"]=election_2012["통합진보당"].apply(lambda x : str(x))
election_2012["독거노인수"]=election_2012["독거노인수"].apply(lambda x : str(x))
election_2012["사업체수"]=election_2012["사업체수"].apply(lambda x : str(x))
election_2012["종사자수"]=election_2012["종사자수"].apply(lambda x : str(x))
election_2012["기초수급자인원수"]=election_2012["기초수급자인원수"].apply(lambda x : str(x))
election_2012["보육시설"]=election_2012["보육시설"].apply(lambda x : str(x))
```

```python
election_2012["선거인수"]=election_2012["선거인수"].replace("nan", 0)
election_2012["새누리당"]=election_2012["새누리당"].replace("nan", 0)
election_2012["민주통합당"]=election_2012["민주통합당"].replace("nan", 0)
election_2012["통합진보당"]=election_2012["통합진보당"].replace("nan", 0)
election_2012["독거노인수"]=election_2012["독거노인수"].replace("nan", 0)
election_2012["사업체수"]=election_2012["사업체수"].replace("nan", 0)
election_2012["종사자수"]=election_2012["종사자수"].replace("nan", 0)
election_2012["기초수급자인원수"]=election_2012["기초수급자인원수"].replace("nan", 0)
election_2012["보육시설"]=election_2012["보육시설"].replace("nan", 0)
```

```python
election_2012["선거인수"]=election_2012["선거인수"].apply(lambda x : int(str(x).replace(",", "")))
election_2012["새누리당"]=election_2012["새누리당"].apply(lambda x : int(str(x).replace(",", "")))
election_2012["민주통합당"]=election_2012["민주통합당"].apply(lambda x : int(str(x).replace(",", "")))
election_2012["통합진보당"]=election_2012["통합진보당"].apply(lambda x : int(str(x).replace(",", "")))
election_2012["독거노인수"]=election_2012["독거노인수"].apply(lambda x : int(str(x).replace(",", "")))
election_2012["사업체수"]=election_2012["사업체수"].apply(lambda x : int(str(x).replace(",", "")))
election_2012["종사자수"]=election_2012["종사자수"].apply(lambda x : int(str(x).replace(",", "")))
election_2012["기초수급자인원수"]=election_2012["기초수급자인원수"].apply(lambda x : int(str(x).replace(",", "")))
election_2012["보육시설"]=election_2012["보육시설"].apply(lambda x : int(str(x).replace(",", "")))
```

```python
election_2012["독거노인수"]=election_2012["독거노인수"].replace(0, election_2012["독거노인수"].median())
election_2012["사업체수"]=election_2012["사업체수"].replace(0, election_2012["사업체수"].median())
election_2012["종사자수"]=election_2012["종사자수"].replace(0, election_2012["종사자수"].median())
election_2012["기초수급자인원수"]=election_2012["기초수급자인원수"].replace(0, election_2012["기초수급자인원수"].median())
election_2012["보육시설"]=election_2012["보육시설"].replace(0, election_2012['보육시설'].median())
```

```python
election_2012["유년부양비"]=election_2012["유년부양비"].fillna(election_2012['유년부양비'].median())
election_2012["노년부양비"]=election_2012["노년부양비"].fillna(election_2012['노년부양비'].median())
election_2012["노령화지수"]=election_2012["노령화지수"].fillna(election_2012['노령화지수'].median())
election_2012["사업체 창업률"]=election_2012["사업체 창업률"].fillna(election_2012["사업체 창업률"].median())
election_2012["금융기관"]=election_2012["금융기관"].fillna(election_2012["금융기관"].median())
election_2012["사업체 평균연령"]=election_2012["사업체 평균연령"].fillna(election_2012["사업체 평균연령"].median())
election_2012["장애등급별 장애인현황"]=election_2012["장애등급별 장애인현황"].fillna(election_2012["장애등급별 장애인현황"].median())
```

```python
election_2012["장애등급별 장애인현황"]=election_2012["장애등급별 장애인현황"].replace(0, election_2012['장애등급별 장애인현황'].median())
```
