```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#가공할 데이터 호출
data = pd.read_csv('https://github.com/lhmlhm1111/Data_GongHak/raw/master/11-22totaldata.txt',encoding='UTF-8',sep=',')
```

```python
#데이터 가공 : 변수 선별 (제외할 변수)
df1 = data.drop(['year','읍면동명','자치구','선거인수','인구'],axis=1)
```

```python
#상관계수 분포 지도 (Heat Map) 작성
plt.rc('font', family='Malgun Gothic')

fig = plt.figure(figsize=[20,20])

ax=sns.heatmap(df1.corr(),annot=True,square=True)
```
